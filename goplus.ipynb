```python
from dune_client.client import DuneClient
dune = DuneClient("JSiwL5HkeaqB8rs9k9mriuBua5D3tCPR")
query_result = dune.get_latest_result(3932992)
query_result
```

```python
import pandas as pd
from datetime import datetime
from dateutil import tz

# 假设这是从 Dune API 获取的输出
query_result = dune.get_latest_result(3932992)

# 提取基本信息
execution_id = query_result.execution_id
query_id = query_result.query_id
state = query_result.state
submitted_at = query_result.times.submitted_at
execution_started_at = query_result.times.execution_started_at
execution_ended_at = query_result.times.execution_ended_at
expires_at = query_result.times.expires_at
cancelled_at = query_result.times.cancelled_at

# 提取结果行并转换为 DataFrame
result_rows = query_result.result.rows if query_result.result.rows else []
df = pd.DataFrame(result_rows)

# 添加额外的元数据到 DataFrame
df['execution_id'] = execution_id
df['query_id'] = query_id
df['state'] = state
df['submitted_at'] = submitted_at
df['execution_started_at'] = execution_started_at
df['execution_ended_at'] = execution_ended_at
df['expires_at'] = expires_at
df['cancelled_at'] = cancelled_at

# 显示 DataFrame
df
```

```python
df.to_csv("goplus_profit.csv")
```

```python
import matplotlib.pyplot as plt

# 确保 day 是日期格式
df['day'] = pd.to_datetime(df['day'])

# 设置图形大小
plt.figure(figsize=(12, 6))

# 计算每日利润的总和
daily_profit = df.groupby('day')['GoPlus_Ecosystem_Profit'].sum()

# 绘制每日利润的条形图
plt.bar(daily_profit.index, daily_profit.values, color='blue', alpha=0.7)

# 添加标题和标签
plt.title('')
plt.xlabel('Date')
plt.ylabel('Daily Ecosystem Profit')

# 显示网格
plt.grid(axis='y')

# 显示图形
plt.xticks(rotation=45)  # 旋转 x 轴标签以便更好地显示
plt.tight_layout()  # 调整布局以适应标签
plt.show()
```
